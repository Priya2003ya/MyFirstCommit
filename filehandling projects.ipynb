```python
import json

def read_data():
    try:
        with open("info.json","r")as file:
            data=json.load(file)
    except FileNotFoundError:
        data=[]
        
    return data
```

```python
def write_data(data):
    with open("info.json","w")as file:
        json.dump(data,file,indent=1)
```

```python
def add_data():
    data=read_data()
    name=input("Enter the name:")
    rollno=int(input("Enter the roll no:"))
    Address=input("Enter the address:")
    sub=input("Enter your subject:")
    Class=int(input("Enter the class:"))
    Sec=input("Enter the section:")
    gmail=input("Enter the gmail:")
    Mo=int(input("Enter the mobile number:"))
    new_data={"Name":name ,"Rollno":rollno ,"Address":Address ,"Subject":sub ,"Class":Class ,"Section":Sec ,"Gmail":gmail ,"MobileNo.":Mo}
    data.append(new_data)
    write_data(data)
    
    print("Data added successfully!")
```

```python
def display_data():
    data=read_data()
    if data:
        print("List of students")
        for i in data:
            print(f"Name:{i['Name']} , Rollno:{i['Rollno']} ,  Address:{i['Address']} ,  Subject:{i['Subject']} ,  Class:{i['Class']} ,  Section:{i['Section']} ,  Gmail:{i['Gmail']} ,  MobileNo.:{i['MobileNo.']}\n")
    else:
        print("Data not found")
            
```

```python
def search_data():
    data=read_data()
    search_in_roll=int(input("Enter the roll number for search data:"))
    for i in data:
        if search_in_roll==i["Rollno"]:
            print(f"Name:{i['Name']} , Address:{i['Address']} ,  Subject:{i['Subject']} ,  Class:{i['Class']} ,  Section:{i['Section']} ,  Gmail:{i['Gmail']} ,  MobileNo.:{i['MobileNo.']}\n")
        else:
            print("No data found")
```

```python
def update_data():
    data=read_data()
    update_to_roll = int(input("Enter the roll number for update data:"))
    for i in data:
        if update_to_roll == i["Rollno"]:
            new_name=input("Enter the ne name:")
            new_mobile=int(input("Enter the new mobile:"))
            new_address=input('Enter the new address:')
            i["Name"]=new_name
            i["MobileNo."]=new_mobile
            i["Address"]=new_address
            write_data(data)
            print("Data updated successfully!")
            return
    print("Data not found.")
```
